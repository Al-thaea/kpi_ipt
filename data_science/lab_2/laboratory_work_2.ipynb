Зчитати завантажені текстові файли у фрейм (
Імена стовбців фрейму мають бути змістовними та легкими л

```python
folder = 'data'
files = os.listdir(folder)
# ids = 0
df_s = pd.DataFrame()

for file_name in files:
    # ids += 1
    
    headers = ['Year', 'Week', 'SMN', 'SMT', 'VCI', 'TCI', 'VHI', 'empty']
    df = pd.read_csv(rf'{folder}\{file_name}', header=1, names=headers, skiprows=1)[:-1]
    df = df.drop(df.loc[df['VHI'] == -1].index)
    df['Area'] = int(file_name.split("_")[2])
    # print(df.head(), '\n')
    # df_all = pd.concat([df_all, df], ignore_index=True)
    df = df.drop(columns=['empty'])
    df_s = df_s.dropna(axis=1, how='all')
    df_s = pd.concat([df_s, df]).drop_duplicates().reset_index(drop=True)
    
print(df_s, '\n')
```

Обрахунок екстремумів для заданої області та року

```python
def vhi_calculate (Year, Area):
    df_new = df_s[(df_s.Area == Area) & (df_s.Year.astype(str) == str(Year))]['VHI']
    print('For ' + str(Year) + ' year and the province ' + str(Area) + ':')
    print('Min: ' + str(df_new.min()))
    print('Max: ' + str(df_new.max()))

    
vhi_calculate(2000, 21)
```

Виведення ряду VHI для області за вказаний рік

```python
def vhi_values(y, area):
    filter_df_s = df_s[(df_s.Year.astype(str) == str(y)) & (df_s.Area == area)]
    vhi_week_values = filter_df_s[['Week', 'VHI']]
    return vhi_week_values.to_string(index=False) #після крапки приховує номер рядка у загальному датасеті

print(vhi_values(2002, 21))

```

Виведення ряду VHI за вказаний діапазон років для вказаних областей

```python
def vhi_range(year_start, year_end, areas):
    # Перетворюємо рік у рядок для порівняння з рядком
    year_start = str(year_start)
    year_end = str(year_end)
    # Фільтруємо датафрейм за діапазоном років та областями
    filtered_df = df_s[(df_s['Year'].astype(str).between(year_start, year_end)) & (df_s['Area'].isin(areas))]
    # Групуємо дані за роком та областю та обчислюємо середнє значення VHI
    grouped_df = filtered_df.groupby(['Year', 'Area'])['VHI'].mean().reset_index()
    return grouped_df

# Приклад виклику функції для отримання середнього значення VHI за діапазоном років 2000-2020 для областей 21 і 22
result = vhi_range(2000, 2005, [21, 22])
print(result)

```

Для всього набору даних виявити роки, протягом яких екстремальні посухи торкнулися більше вказаного відсотка областей по Україні (20% областей - 5 областей з 25). Повернути роки, назви областей з екстремальними посухами та значення VHI.

Класифікація посухи за значеннями VHI:
- VHI < 15 – посуха, інтенсивність якої від середньої до надзвичайної. - екстремальні посухи
- VHI < 35 – посуха, інтенсивність якої від помірної до надзвичайної. - помірні посухи


```python
def extreme_drought():
    df_s_drought = df_s[df_s['VHI'] <= 15][['VHI', 'Year', 'Week', 'Area']]
    df_s_drought = df_s_drought.drop_duplicates(subset=['Year', 'Area'])
    grouped_areas = df_s_drought.groupby('Year')['Area'].nunique()
    if grouped_areas.empty:
        print("В Україні не було екстремальних посух")
        return
        
extreme_drought()
```

```python
def medium_drought(amount):
    df_s_drought = df_s[df_s['VHI'] <= 35][['VHI', 'Year', 'Week', 'Area']]
    df_s_drought = df_s_drought.drop_duplicates(subset=['Year', 'Area'])
    grouped_areas = df_s_drought.groupby('Year')['Area'].nunique()
    for year, area_count in grouped_areas.items():
        if area_count > amount:
            print("Рік:", year)
            print(df_s_drought[df_s_drought['Year'] == year])

medium_drought(5)


```
